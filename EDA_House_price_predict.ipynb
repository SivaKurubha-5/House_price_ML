# EDA (Exploratory Data Analysis)

## Univarite analysis

```python
# title column

font_title = {
    'family': 'Times New Roman',
    'size': 12,
    'weight': 'bold',
    'color': 'black'
}
plt.figure(figsize = (5,4))
ax = sns.countplot(data,x = 'Title',palette = 'colorblind')
ax.set_axisbelow(True)
for i in ax.patches:
    h = i.get_height()
    ax.text(i.get_x() + i.get_width()/2.,h+3,f'{int(h)}',ha = 'center',fontsize = 9)
plt.title("Individual Groups count",fontdict = font_title)
plt.xticks(rotation = 90)
plt.xlabel("Title",fontdict = font_title)
plt.ylabel("Frequency",fontdict = font_title)
plt.ylim(0,30000)
plt.grid(True,axis = "y")
plt.show()
```

```python
#'Amount(in rupees)'

plt.figure(figsize = (5,3))

ax = sns.scatterplot(data,x ='Amount(in rupees)',y = 'Amount(in rupees)' )
ax.set_axisbelow(True)
plt.title("Amount Distribution",fontdict = font_title)
plt.xlabel("Amount",fontdict = font_title)
plt.xticks(rotation = 90)
plt.ylabel('Frequency',fontdict = font_title)
plt.ticklabel_format(style= "plain")
plt.grid(True)
plt.show
```

## Bivariate Analysis

```python
# Title vs Amount

formatter = ScalarFormatter()
ax = data.groupby('Title')['Amount(in rupees)'].mean().plot(kind = 'bar')
ax.set_axisbelow(True)
for i in ax.patches:
    h = i.get_height()
    ax.text(i.get_x() + i.get_width()/2.,h+3,f'{int(h)}',ha = 'center',fontsize = 9)
plt.xlabel("Title",fontdict = font_title)
plt.ylabel('Frequency',fontdict = font_title) 
plt.title("Homes Average Amount",fontdict = font_title)
formatter.set_scientific(False)  # Disable scientific notation
formatter.set_useOffset(False) 
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True,axis = 'y')
plt.show()
```

```python
# Top 10 locations largely used Apartments

bar = data[data['Location'].isin(data['Location'].value_counts()[:10].index)]
bar = pd.DataFrame(bar.groupby(['Location','Title'])['Title'].count().sort_values(ascending = False)[:10])
bar.columns = ['Count']

ax = sns.barplot(bar,x = bar.index.get_level_values(0),y = 'Count',hue = bar.index.get_level_values(1))
ax.set_axisbelow(True)
for i in ax.patches:
    h = i.get_height()
    ax.text(i.get_x() + i.get_width()/2.,h+3,f'{int(h)}',ha = 'center',fontsize = 9)
plt.xlabel("Location",fontdict = font_title)
plt.xticks(rotation = 90)
plt.ylabel('Frequency',fontdict = font_title) 
plt.title("Locations top selling Aparments",fontdict = font_title)
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True,axis = 'y')
plt.show()
```

```python
# Apartments avg carpet Area

bar =  data.groupby('Title')['Carpet_Area'].mean()
ax = sns.barplot(bar,palette = 'deep')
ax.set_axisbelow(True)
for i in ax.patches:
    h = i.get_height()
    ax.text(i.get_x() + i.get_width()/2.,h+3,f'{int(h)}',ha = 'center',fontsize = 9)
plt.xlabel("Title",fontdict = font_title)
plt.xticks(rotation = 90)
plt.ylabel('Frequency',fontdict = font_title) 
plt.title("Apartments Average Carpet Areas",fontdict = font_title)
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True,axis = 'y')
plt.show()
```

```python
# Each Apartment Statues

bar = pd.DataFrame(data.groupby(['Title','Status'])['Status'].count())
bar.columns = ['Count']
ax = sns.barplot(bar,x = bar.index.get_level_values(0),y = 'Count',hue = bar.index.get_level_values(1))
ax.set_axisbelow(True)
plt.xlabel("Title",fontdict = font_title)
plt.xticks(rotation = 90)
plt.ylabel('Frequency',fontdict = font_title) 
plt.title("Apartments Statues",fontdict = font_title)
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True,axis = 'y')
plt.show()
plt.show()
```
