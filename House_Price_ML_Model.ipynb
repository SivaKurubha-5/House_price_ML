```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
from sklearn.preprocessing import StandardScaler
```

```python
Data =pd.read_csv("Clean_House_price_data.csv")
data = Data.copy()
```

```python
data.head()
```

```python
data.shape
```

__As per Indian survery compaines(Knight Frank India,NoBroker survey,..etc) indian are willing to buy houses price upto 3 cr NRI's upt o 100cr__

1.Dealing with these people price column outliers are removed fixed max amount(100 cr)

```python
data = data[data['Amount(in rupees)']<=100000000]
```

```python
data.shape
```

## Data Encoding

### Label Encoding

```python
# Label encoding furnishing column

Furnishing_values = {'Unfurnished' : 1,
                    'Semi-Furnished' : 2,
                    'Furnished' : 3,
                    'Not-Mentioned' : 0}
```

```python
data['Furnishing'] = data['Furnishing'].map(Furnishing_values)
```

```python
# Facing column label encoding

Facing_values = {'0' : 0,
                 'South -West' : 1,
                 'South' : 3,
                 'North - West' : 4,
                 'South - East' : 5,
                 'West' : 6, 
                 'North' : 8,
                 'East': 9 ,
                 'North - East' : 10 ,
                }
```

```python
data['Facing'] = data['Facing'].map(Facing_values)
```

```python
data.head()
```

```python
data.info()
```

```python
cat_data = data.select_dtypes(include = 'object')
num_data = data.select_dtypes(include = 'number')
```

```python
cat_data.info()
```

```python
num_data.info()
```

```python
one_hot_cat = pd.get_dummies(cat_data,dtype = 'int')
```

```python
one_hot_cat
```

```python
to_be_scaled = pd.concat([num_data,one_hot_cat],axis = 1)
```

```python
to_be_scaled
```

```python
scaler = StandardScaler()
scaled_all = scaler.fit_transform(to_be_scaled)
scaled_df = pd.DataFrame(scaled_all, columns=to_be_scaled.columns)
```

```python
scaled_df.shape
```

```python
y = scaled_df['Amount(in rupees)']
x = scaled_df.drop('Amount(in rupees)',axis = 1)
```

```python
x.shape,y.shape
```

# ML Linear regression Model

```python
# importing Machine learning models

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# spliting train and test data

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.10,random_state= 99)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
# model intialization

model =  LinearRegression() 
```

```python
# model training

model.fit(x_train,y_train)
```

```python
# model testing

x_test_predict = model.predict(x_test)
x_train_predict = model.predict(x_train)
```

```python
# train mean_squared_error

print("train mean squared error : ",mean_squared_error(y_train,x_train_predict))
```

```python
# test means squared error

print("test mean squared error : " ,mean_squared_error(y_test,x_test_predict))
```

```python
# r2 score of train 

print("r2 score",r2_score(y_train,x_train_predict))
```

```python
# r2 score of test

print("r2 score",r2_score(y_test,x_test_predict))
```

### Ridge

```python
# ridge intialization

ridge_model = Ridge(alpha= 0.001)
ridge_model.fit(x_train,y_train)
```

```python
ridge_train_predict  = ridge_model.predict(x_train)
ridge_test_predict = ridge_model.predict(x_test)
```

```python
# ridge mean squared error train
print("ridge mean squared error : ",mean_squared_error(y_train,ridge_train_predict))
```

```python
# ridge mean squared error of test
print("ridge mean squared error : ",mean_squared_error(y_test,ridge_test_predict))
```

```python
# r2 scores of train 
print("R2 scores or train : ",r2_score(y_train,ridge_train_predict))
```

```python
#r2 score of test 
print("r2 socres of test : ",r2_score(y_test,ridge_test_predict))
```

### lasso

```python

Lasso_model = Lasso(alpha = 0.001)
Lasso_model.fit(x_train,y_train)
```

```python
Lasso_train_predict = Lasso_model.predict(x_train)
Lasso_test_predict = Lasso_model.predict(x_test)
```

```python
# mean squared error of x_train predict 
print("mean sqaured error of train  : ",mean_squared_error(y_train,Lasso_train_predict))
```
